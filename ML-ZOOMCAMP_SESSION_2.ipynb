```python
import pandas as pd
import numpy as np
```

```python
data = 'https://www.kaggle.com/CooperUnion/cardataset/download'
```

```python
import wget
```

```python
wget.download(data, 'car_data.csv')
```

```python
df = pd.read_csv('car_data.csv')
```

```python
df.head()
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.dtypes[df.dtypes == 'object']
```

```python
strings = list(df.dtypes[df.dtypes == 'object'].index)
```

```python
strings
```

```python
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ', '_')
```
